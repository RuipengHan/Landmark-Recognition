```python
# prompt: I have a csv "label_csv = "/content/drive/MyDrive/cs444-final-project/project/data/train/selected_train.csv" that has three columns. The third one is the "landmark_id" which has a total of 2000 unique values (that are nonrelevant discrete integers, for example, 1231, 43388, 213002). I want to transform the column so that each of the values take on a value from 0 to 1999

import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(label_csv)

# Get the unique values in the "landmark_id" column
unique_landmarks = df["landmark_id"].unique()
print(len(unique_landmarks))
```

```python

# Create a dictionary to map the unique values to new values from 0 to 1999
landmark_to_index = {landmark: i for i, landmark in enumerate(unique_landmarks)}

# Apply the mapping to the "landmark_id" column
df["landmark_id_transformed"] = df["landmark_id"].apply(lambda x: landmark_to_index[x])

# Save the modified DataFrame back to the CSV file
df.to_csv(label_csv, index=False)
```

```python
import os
# Path to the directory containing imagesfolder_path ='/content/drive/MyDrive/data/selected_train'
# Image file extension to check
image_extension ='.jpg'
image_files = []
folder_path = "/content/drive/MyDrive/cs444-final-project/project/data/selected_train"
with os.scandir(folder_path) as entries:
  for entry in entries:
    if entry.is_file() and entry.name.lower().endswith(image_extension):
      # print(entry.name)
      image_files.append(entry.name)
      # break


```

```python
# prompt: I want to save image_files, which is a list of strings, into a single column (named "id") csv; the order of the rows should be preserved as the order of the image_files list.

import pandas as pd

image_files_df = pd.DataFrame(image_files, columns=["id"])

image_files_df.to_csv("/content/drive/MyDrive/cs444-final-project/project/data/train_file_names.csv", index=False)
```
